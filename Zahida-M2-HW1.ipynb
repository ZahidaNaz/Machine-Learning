```python
missing=["A","B","Others","ABC",45]
dataset=pd.read_csv('Cricket.csv',na_values=missing)
```

```python
median=dataset['ConsecutiveWins'].median()
dataset['ConsecutiveWins'].fillna(median,inplace=True)
```

```python
dataset['Opposition_Others'].mode()
dataset['Opposition_Others'].fillna(dataset['Opposition_Others'].mode()[0],inplace=True)
```

```python
dataset['Country'].mode()
dataset['Country'].fillna(dataset['Country'].mode()[0],inplace=True)
```

```python
dataset['Home_Ground'].mode()
dataset['Home_Ground'].fillna(dataset['Home_Ground'].mode()[0],inplace=True)
```

```python
dataset['BattingFirst'].mode()
dataset['BattingFirst'].fillna(dataset['BattingFirst'].mode()[0],inplace=True)
```

```python
dataset['Pak_Win'].mode()
dataset['Pak_Win'].fillna(dataset['Pak_Win'].mode()[0],inplace=True)
```

```python
dataset.rename(columns={ dataset.columns[9]: "Combination_Changed" },inplace=True)
```

```python
dataset['Combination_Changed'].mode()
dataset['Combination_Changed'].fillna(dataset['Combination_Changed'].mode()[0],inplace=True)
```
